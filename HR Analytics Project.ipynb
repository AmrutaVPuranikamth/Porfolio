# Attrition by department

```python
# Filter the data to show 'Yes' values in 'Attrition' column
attrition_data= data[data['Attrition']=='Yes']

# Calculate the count of attrition by department
attrition_dpt=attrition_data.groupby(['Department']).size().reset_index(name='Count')

#Create a bar graph
labels=list(attrition_dpt['Department'])
values=list(attrition_dpt['Count'])

fig = plt.figure(figsize=(8,4))
plt.bar(labels, values,color='green',
         width=0.2)

plt.xlabel('Department')
plt.ylabel('No. of Employees')
plt.title("Attrition by Department")


#Show the graph
plt.show()
```

```python
#Conclusion: The Research & Development department has a high attrition rate.
```

# Attrition by education field

```python
attrition_dpt= attrition_data.groupby(['EducationField']).size().reset_index(name='Count')

#Create a bar plot
labels= list(attrition_dpt['EducationField'])
values=list(attrition_dpt['Count'])

fig = plt.figure(figsize = (10,5))
plt.bar(labels, values, color='green',
         width=0.4)

plt.xlabel('EducationField')
plt.ylabel('No. of Employees')
plt.title('Attrition by Education Field')

#Show the chart
plt.show()
```

```python
#Conclusion: Employees with Life Sciences as an education field have a high attrition rate.
```

# Attrition by number of years since last promotion

```python
attrition_dpt= attrition_data.groupby(['YearsAtCompany']).size().reset_index(name='Count')

#Create a bar graph
labels= list(attrition_dpt['YearsAtCompany'])
values= list(attrition_dpt['Count'])

fig = plt.Figure(figsize =(12,6))
plt.bar(labels, values, color='green',
         width=0.8)

plt.xlabel('YearsAtCompany')
plt.ylabel('No. of Employees')
plt.title('Attrition by number of years since last promotion')

#Show the chart
plt.show()
```

```python
#Conclusion: Employees who dont get promotions leave organization compared to the employees who got promotions.
```

# Attrition by gender

```python
attrition_dpt=attrition_data.groupby(['Gender']).size().reset_index(name='Count')

# Create a donut chart
fig= go.Figure(data=[go.Pie(
     labels=attrition_dpt['Gender'],
     values=attrition_dpt['Count'],
     hole=0.4,
     marker=dict(colors=['green', 'maroon']),
     textposition='inside'
)])

# Update the layout
fig.update_layout(title='Attrition by Gender', font=dict(size=16), legend=dict(
    orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1
))

# Show the chart
fig.show()
```

```python
#Conclusion: Men have high attrition rate compared to women.
```

# Attrition rate between monthly income and the age of the employees

```python
fig =px.scatter(data, x='Age', y='MonthlyIncome', color='Attrition', trendline='ols')
fig.update_layout(title='Age vs. Monthly Income by Attrition')
fig.show()
```

```python
#Conclusion: As we can see the age of the person increases, monthly income also increases. We can see a high rate of attrition among the employees with low monthly incomes
```
